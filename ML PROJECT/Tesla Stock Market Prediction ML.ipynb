```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
```

```python
Tesla.plot.line(y="Close", use_index=True)
```

```python
# cleaning the data

del Tesla["Dividends"]
del Tesla["Stock Splits"]
```

```python
# predicting the stock

Tesla["Tomorrow"] = Tesla["Close"].shift(-1)
```

```python
Tesla
```

```python
Tesla["Target"] = (Tesla["Tomorrow"] > Tesla["Close"]).astype(int)
```

```python
Tesla
```

```python
# Training an initial Machine Learning Model

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = Tesla.iloc[:-100]
test = Tesla.iloc[-100:]

predictors = ["Close", "Volume", "Open", "High", "Low"]
model.fit(train[predictors], train["Target"])
```

```python
from sklearn.metrics import precision_score

preds = model.predict(test[predictors])
```

```python
preds = pd.Series(preds, index=test.index)
```

```python
precision_score(test["Target"], preds)
```

```python
combined = pd.concat([test["Target"], preds], axis=1)
```

```python
combined.plot()
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
# Backtesting to train the model using 5 years data

def backtest(data, model, predictors, start=1250, step=250):
    all_predictions = []
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[:i].copy()  # Use i directly instead of 0+i
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(Tesla, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
# Adding additional predictors to our model

horizons = [2, 5, 60, 250, 1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = Tesla['Close'].rolling(window=horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    Tesla.loc[:, ratio_column] = Tesla["Close"] / rolling_averages
    
    trend_column = f"Trend_{horizon}"
    Tesla.loc[:, trend_column] = Tesla['Target'].shift(1).rolling(window=horizon).sum()
    
    new_predictors.extend([ratio_column, trend_column])
```

```python
Tesla = Tesla.dropna()
```

```python
Tesla
```

```python
# improving the model

def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors]) [:,1]
    preds[preds >= .6] = 1
    preds[preds < .6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(Tesla, model, new_predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```
